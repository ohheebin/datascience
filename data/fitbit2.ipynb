```python
#데이터 추가하는 파이썬
import pandas as pd
import datetime
import csv
import numpy as np
import matplotlib.image
import matplotlib.pyplot as plt
import gmaps
import gmaps.datasets
```

```python
step = pd.read_csv('./step.csv')
step
```

```python
step['weekday'] = pd.DatetimeIndex(step['date']).weekday
step['weekday_name'] = pd.DatetimeIndex(step['date']).weekday_name
step['year'] = pd.DatetimeIndex(step['date']).year
step['month'] = pd.DatetimeIndex(step['date']).month
step['day'] = pd.DatetimeIndex(step['date']).day
step['hour'] = pd.DatetimeIndex(step['date']).hour
step['week'] = pd.DatetimeIndex(step['date']).week
step
```

```python
df = pd.DataFrame(step)
filename = 'result_step.csv'
df.to_csv(filename, index=False, encoding='utf-8')
```

```python
steps = pd.read_csv('./step_15mins.csv')
steps
```

```python
steps['weekday'] = pd.DatetimeIndex(steps['date']).weekday
steps['weekday_name'] = pd.DatetimeIndex(steps['date']).weekday_name
steps['year'] = pd.DatetimeIndex(steps['date']).year
steps['month'] = pd.DatetimeIndex(steps['date']).month
steps['day'] = pd.DatetimeIndex(steps['date']).day
steps['hour'] = pd.DatetimeIndex(steps['date']).hour
steps['week'] = pd.DatetimeIndex(steps['date']).week
steps['minute'] = pd.DatetimeIndex(steps['date']).minute
steps['HM'] = steps['hour']*60 + steps['minute']
```

```python
df = pd.DataFrame(steps)
filename = 'result_step_15min.csv'
df.to_csv(filename, index=False, encoding='utf-8')
```
